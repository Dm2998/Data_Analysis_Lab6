## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
x = np.arange(0, 50, 50)
y = x * 2

# Create a figure object and two axes
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

# Plot (x, y) on both axes
ax1.plot(x, y)
ax2.plot(x, y)

# Show the plot
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))


ax1.plot(x, y, color='blue', linestyle='--', label='(x, y)')


ax2.plot(x, z, color='red', lw=3)


plt.tight_layout()

plt.show()
```
